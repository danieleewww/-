```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
T=20
dt=0.01
N=round(T/dt)
t=np.linspace(0,T,N)
```

```python
z=np.random.standard_normal(size=N)
```

```python
z=np.cumsum(z)*np.sqrt(dt)
```

```python
plt.xlabel('t')
plt.ylabel('z')
plt.title(u'BM')
plt.plot(t,z)
plt.show()
```

```python
a=0.1
b=0.6
x=a*t+b*z
```

```python
plt.xlabel('t')
plt.ylabel('x')
plt.title(u'BI')
plt.plot(t,x)
plt.show()
```

```python
s0=0.01
s=s0*np.exp(x)
```

```python
plt.xlabel('t')
plt.ylabel('s')
plt.title(u'GBM')
plt.plot(t,s)
plt.show()
```
